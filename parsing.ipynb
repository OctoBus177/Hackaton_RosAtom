```python
def get_free_proxies():
    url = "http://free-proxy-list.net"
    tmp = requests.get(url)

    #   HTTP    soup
    soup = BeautifulSoup(tmp.content, "html.parser")
    proxies = []
    soup_body = soup.body
    arr = []
    table = soup_body.find_all('tr')
 
    for i in table:
        res = i.find_all('td')
        count = 0
        r = ''
        for j in res:

            tex = j.text    #print(res,'\n\n\n\n')
            if count == 0:
                r = str(tex)
            elif count == 1:
                r = r + ':' + str(tex)
            
            if (count > 2) and (tex.isdigit()):
                continue
            count +=1
        if count == 8:
            arr.append(r)
    return arr
```

```python
def get_session(proxies):
    #  HTTP
    session = requests.Session()
    #    
    proxy = random.choice(proxies)
    session.proxies = {"http": proxy, "https": proxy}
    return session
```

```python
proxies = get_free_proxies()
for i in range(5):
    s = get_session(proxies)
    try:
        print("   IP:", s.get("http://icanhazip.com", timeout=1.5).text.strip())
    except Exception as e:
        continue
```
